```python
import pandas as pd

chunk_size = 100
count = 1
citi_lei = "6SHGI4ZSSLCXXQSBB395"
df_short = None
file_path = './20240430-0000-gleif-goldencopy-lei2-golden.csv'
# Define your condition function
def citi_condition(chunk):
  # Apply your condition to the chunk
  return chunk[chunk['LEI'] == citi_lei]

for chunk in pd.read_csv(file_path, chunksize=chunk_size):
  df1 = citi_condition(chunk)
  
  if not df1.empty:
    df_short = chunk
    break
```

```python
df_short1 = df_short.dropna(how='all', axis=1, inplace=False)
```
