```python
import pandas as pd

chunk_size = 1000
count = 1
citi_lei = "6SHGI4ZSSLCXXQSBB395"
df_short = pd.DataFrame()
file_path = './20240502-0000-gleif-goldencopy-rr-golden.csv'
# Define your condition function
def citi_condition(chunk):
  # Apply your condition to the chunk
  return chunk[chunk['Relationship.EndNode.NodeID'] == citi_lei]

dataframe = pd.read_csv(file_path, encoding='utf', chunksize=chunk_size) #TextFileReader - output type format

for chunk in dataframe:
   df1 = citi_condition(chunk)
   if not df1.empty:
      for i, chunk in df1.iterrows():
        if chunk['Relationship.EndNode.NodeID'] == citi_lei:
          count = count + 1
          df_short = df_short.add(chunk)
          df_short.loc[len(df_short.index)] = chunk
print(count)
```

```python
df_short = df_short.dropna(how='all', axis=1, inplace=False)
```
